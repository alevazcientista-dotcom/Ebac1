**1 - Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.**

```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
#%matplotlib notebook
```

```python
display(sinasc.describe())
```

```python
# Carregar a base de dados
sinasc = pd.read_csv('/content/M6 - SINASC_RO_2019.csv')
```

```python
# Configurar estilo dos gráficos
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('crest')
```

```python
print("Colunas do DataFrame sinasc:")
display(sinasc.columns)
```

```python
print("Distribuição dos valores na coluna 'SEXO':")
display(sinasc['SEXO'].value_counts())
```

```python
# boxplot da idade da mãe
plt.figure(figsize=(6,4))
plt.boxplot(sinasc['IDADEMAE'].dropna())
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade da Mãe')
plt.show()
```

```python
#Boxplot da idade da mãe
plt.figure(figsize=(7,5))
sns.boxplot(x=sinasc['IDADEMAE'], color='skyblue')
plt.title('Boxplot da Idade da Mãe', fontsize=14)
plt.xlabel('Idade da Mãe')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Verificar rapidamente as primeiras linhas
print("Visualizando as primeiras linhas do dataset:")
display(sinasc.head())
```

**2 - Faça um boxplot do peso do bebê**

```python
#boxplot do peso bebê
plt.figure(figsize=(6,4))
plt.boxplot(sinasc['PESO'].dropna())
plt.title('Boxplot do Peso do Bebê')
plt.ylabel('Peso (gramas)')
plt.show()
```

**3 - Faça um histograma da idade da mãe**

```python
# Histograma da idade da mãe
plt.figure(figsize=(7,5))
sns.histplot(sinasc['IDADEMAE'].dropna(), bins=20, kde=True, color='royalblue')
plt.title('Histograma da Idade da Mãe', fontsize=14)
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**4 - Faça um histograma do peso do bebê**

```python
# Histograma do Peso do bebê
plt.figure(figsize=(7,5))
sns.histplot(sinasc['PESO'].dropna(), bins=20, kde=True, color='mediumseagreen')
plt.title('Histograma do Peso do Bebê', fontsize=14)
plt.xlabel('Peso do Bebê (gramas)')
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
