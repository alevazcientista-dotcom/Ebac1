```python
# Pacotes principais
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurações visuais
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')
```

```python
# Caminho do arquivo (ajuste se necessário)
sinasc = pd.read_csv('/content/M6 - SINASC_RO_2019.csv')
```

**2 - Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.**

```python
print(sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']].dtypes)
```

```python
# Create bar plots for APGAR1, APGAR5, SEXO, GESTACAO, and ESTCIVMAE
variables_to_plot = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

plt.figure(figsize=(15, 10))

for i, var in enumerate(variables_to_plot):
    plt.subplot(2, 3, i + 1) # Arrange plots in a 2x3 grid
    sns.countplot(data=sinasc, x=var, palette='viridis')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent overlapping titles and labels
plt.show()
```

**4 - Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS**

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.countplot(ax=axes[0, 0], data=sinasc, x='Sexo', palette='mako')
axes[0, 0].set_title('Distribuição por Sexo do Bebê')

sns.countplot(ax=axes[0, 1], data=sinasc, x='Gestação', palette='crest')
axes[0, 1].set_title('Duração da Gestação')

sns.countplot(ax=axes[1, 0], data=sinasc, x='Estado Civil Mãe', palette='flare')
axes[1, 0].set_title('Estado Civil da Mãe')

sns.countplot(ax=axes[1, 1], data=sinasc, x='Consultas', palette='viridis') # Corrected column name
axes[1, 1].set_title('Número de Consultas Pré-Natal')

# Ajustar layout
for ax in axes.flat:
    ax.set_xlabel('')
    ax.set_ylabel('Frequência')
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('Painel de Distribuições - SINASC RO 2019', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
column_translation_dict = {
    'SEXO': 'Sexo',
    'GESTACAO': 'Gestação',
    'ESTCIVMAE': 'Estado Civil Mãe',
    'CONSULTAS': 'Consultas',
    'APGAR1': 'APGAR1',
    'APGAR5': 'APGAR5'
}
```

```python
sinasc.rename(columns=column_translation_dict, inplace=True)
sinasc.head()
```

```python
sexo_translation_dict = {
    'Masculino': 'Masculino',
    'Feminino': 'Feminino'
}

# Inspect unique values in 'Gestação' before creating the dictionary
print(sinasc['Gestação'].unique())

gestacao_translation_dict = {
    '22 a 27 semanas': '22 a 27 semanas',
    '32 a 37 semanas': '32 a 37 semanas',
    '38 a 41 semanas': '38 a 41 semanas',
    '42 semanas e mais': '42 semanas e mais',
    '28 a 31 semanas': '28 a 31 semanas',
    'Menos de 22 semanas': 'Menos de 22 semanas',
    'Ignorado': 'Ignorado'
}
```

```python
import numpy as np

gestacao_translation_dict = {
    '37 a 41 semanas': '37 a 41 semanas',
    '42 semanas e mais': '42 semanas e mais',
    '32 a 36 semanas': '32 a 36 semanas',
    '22 a 27 semanas': '22 a 27 semanas',
    '28 a 31 semanas': '28 a 31 semanas',
    'Menos de 22 semanas': 'Menos de 22 semanas',
    np.nan: 'Ignorado' # Include the missing value
}
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=sinasc, x='Sexo', palette='mako')
plt.title('Distribuição por Sexo do Bebê')
plt.xlabel('Sexo')
plt.ylabel('Frequência')
plt.show()
```

```python
sinasc['Sexo'] = sinasc['Sexo'].map(sexo_translation_dict)
sinasc['Gestação'] = sinasc['Gestação'].map(gestacao_translation_dict)
sinasc.head()
```

```python
variables_to_plot = ['APGAR1', 'APGAR5', 'Sexo', 'Gestação', 'Estado Civil Mãe']

plt.figure(figsize=(15, 10))

for i, var in enumerate(variables_to_plot):
    plt.subplot(2, 3, i + 1) # Arrange plots in a 2x3 grid
    if sinasc[var].dtype == 'object':
        sns.countplot(data=sinasc, x=var, palette='viridis')
    else:
        # For numerical APGAR scores, use value_counts and plot bar
        sinasc[var].value_counts().sort_index().plot(kind='bar', color='skyblue')


    # Update titles and labels based on the variable
    if var == 'APGAR1':
        plt.title('Distribuição do APGAR1')
        plt.xlabel('APGAR no 1º minuto')
    elif var == 'APGAR5':
        plt.title('Distribuição do APGAR5')
        plt.xlabel('APGAR no 5º minuto')
    elif var == 'Sexo':
        plt.title('Distribuição por Sexo do Bebê')
        plt.xlabel('Sexo')
    elif var == 'Gestação':
        plt.title('Duração da Gestação')
        plt.xlabel('Semanas de Gestação')
    elif var == 'Estado Civil Mãe':
        plt.title('Estado Civil da Mãe')
        plt.xlabel('Estado Civil')

    plt.ylabel('Frequência')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent overlapping titles and labels
plt.show()

# Plot APGAR5 Médio por Consultas
plt.figure(figsize=(8, 5))
sns.barplot(
    data=sinasc,
    x='Consultas',
    y='APGAR5',
    estimator='mean',
    errorbar=None,
    palette='coolwarm'
)
plt.title('APGAR5 Médio por Número de Consultas Pré-Natal')
plt.xlabel('Número de Consultas')
plt.ylabel('Média do APGAR5')
plt.show()


# Plot panel of 4 graphs
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.countplot(ax=axes[0, 0], data=sinasc, x='Sexo', palette='mako')
axes[0, 0].set_title('Distribuição por Sexo do Bebê')
axes[0, 0].set_xlabel('Sexo')


sns.countplot(ax=axes[0, 1], data=sinasc, x='Gestação', palette='crest')
axes[0, 1].set_title('Duração da Gestação')
axes[0, 1].set_xlabel('Semanas de Gestação')


sns.countplot(ax=axes[1, 0], data=sinasc, x='Estado Civil Mãe', palette='flare')
axes[1, 0].set_title('Estado Civil da Mãe')
axes[1, 0].set_xlabel('Estado Civil')


sns.countplot(ax=axes[1, 1], data=sinasc, x='Consultas', palette='viridis')
axes[1, 1].set_title('Número de Consultas Pré-Natal')
axes[1, 1].set_xlabel('Número de Consultas')


# Ajustar layout
for ax in axes.flat:
    ax.set_ylabel('Frequência')
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('Painel de Distribuições - SINASC RO 2019', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
