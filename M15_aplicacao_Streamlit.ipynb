```python
%%writefile -a app.py
import streamlit as st
import datetime
import time

st.markdown("----")
st.header("Status Elements")

st.subheader("st.progress()")
my_bar = st.progress(0)
for percent_complete in range(100):
    time.sleep(0.05)
    my_bar.progress(percent_complete + 1)

st.subheader("st.spinner()")
with st.spinner('Aguarde...'):
    time.sleep(3)
st.success('Pronto!')

st.subheader("st.balloons()")
if st.button('Explodir Balões!'):
    st.balloons()

st.subheader("st.snow()")
if st.button('Fazer Nevar!'):
    st.snow()
```

```python
%%writefile app.py
import streamlit as st
import datetime
import pandas as pd
import time
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

st.title('Bem vindo')
st.header('Este é um cabeçalho')
st.subheader('Este é um subcabeçalho')
st.write('Este é um texto escrito com st.write().')
st.text('Este é um texto simples com st.text().')
st.caption('Esta é uma legenda.')

st.markdown("----")

st.markdown("Markdown")
st.markdown("_Markdown itálico_")
st.markdown("*Markdown itálico*")
st.markdown("__Markdown negrito__")
st.markdown("**Markdown negrito**")
st.markdown("__*Markdown itálico e negrito*__")

st.markdown("Isso é um link pro [google](https://google.com.br)")

st.markdown("<h1 style='text-align: center; color: red;'>Título em vermelho</h1>", unsafe_allow_html=True)

st.markdown('Olá, mundo :sunglasses:')
st.markdown('Olá, mundo :100:')

st.markdown('''MARKDOWN | PEQUENO | GRANDE
                ---     | ---     | ---
              *iTÁLICO* | `CODE`  | **NEGRITO**
            ''')

st.markdown("----")
st.header("Input Widgets")

text_input_value = st.text_input('Digite seu nome:', 'Seu nome')
st.write('Nome digitado:', text_input_value)

number_input_value = st.number_input('Digite sua idade:', min_value=0, max_value=120, value=30)
st.write('Idade digitada:', number_input_value)

checkbox_value = st.checkbox('Marque se concorda com os termos')
st.write('Checkbox marcado:', checkbox_value)

radio_value = st.radio(
    "Escolha sua fruta favorita:",
    ('Maçã', 'Banana', 'Laranja'))
st.write('Sua fruta favorita é:', radio_value)

selectbox_value = st.selectbox(
    'Selecione sua cor favorita:',
    ('Azul', 'Verde', 'Vermelho'))
st.write('Sua cor favorita é:', selectbox_value)

multiselect_value = st.multiselect(
    'Selecione seus hobbies:',
    ['Ler', 'Escrever', 'Caminhar', 'Cozinhar'])
st.write('Seus hobbies selecionados:', multiselect_value)

slider_value = st.slider('Selecione um valor:', 0, 100, 25)
st.write('Valor selecionado:', slider_value)

date_input_value = st.date_input('Selecione uma data:', datetime.date(2023, 1, 1))
st.write('Data selecionada:', date_input_value)

st.markdown("----")
st.header("Data Display Elements")

# Create a sample DataFrame
data = {'col1': [1, 2, 3, 4],
        'col2': ['A', 'B', 'C', 'D']}
df = pd.DataFrame(data)

st.subheader("st.dataframe()")
st.dataframe(df)

st.subheader("st.table()")
st.table(df)

st.subheader("st.json()")
json_data = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}
st.json(json_data)

st.markdown("----")
st.header("Media Elements")

st.subheader("st.image()")
st.image("https://www.python.org/static/community_logos/python-logo-master-v3-TM.png", caption="Python Logo")

st.subheader("st.audio()")
# Replace with a URL or local path to an audio file
st.audio("https://file-examples-com.github.io/uploads/2017/11/file_example_WAV_1MG.wav")

st.subheader("st.video()")
# Replace with a URL or local path to a video file
st.video("https://file-examples-com.github.io/uploads/2017/04/file_example_MP4_480_1_5MG.mp4")

st.markdown("----")
st.header("Layout and Containers")

st.subheader("st.sidebar()")
with st.sidebar:
    st.write("Este é um texto na barra lateral.")
    sidebar_slider = st.slider('Slider na barra lateral:', 0, 100, 50)
    st.write('Valor do slider na barra lateral:', sidebar_slider)

st.subheader("st.columns()")
col1, col2 = st.columns(2)

with col1:
    st.write("Conteúdo na Coluna 1")
    st.button("Botão na Coluna 1")

with col2:
    st.write("Conteúdo na Coluna 2")
    st.checkbox("Checkbox na Coluna 2")

st.subheader("st.expander()")
with st.expander("Clique para expandir"):
    st.write("Este conteúdo está dentro do expander.")
    st.write("Você pode colocar qualquer elemento Streamlit aqui dentro.")

st.markdown("----")
st.header("Status Elements")

st.subheader("st.progress()")
my_bar = st.progress(0)
for percent_complete in range(100):
    time.sleep(0.05)
    my_bar.progress(percent_complete + 1)

st.subheader("st.spinner()")
with st.spinner('Aguarde...'):
    time.sleep(3)
st.success('Pronto!')

st.subheader("st.balloons()")
if st.button('Explodir Balões!'):
    st.balloons()

st.subheader("st.snow()")
if st.button('Fazer Nevar!'):
    st.snow()

st.markdown("----")
st.header("Chart Elements")

# Create sample data for charts
chart_data = pd.DataFrame(
    np.random.randn(20, 3),
    columns=['a', 'b', 'c'])

st.subheader("Matplotlib Chart")
fig, ax = plt.subplots()
ax.scatter(chart_data['a'], chart_data['b'])
st.pyplot(fig)

st.subheader("Plotly Express Chart")
fig2 = px.line(chart_data, y="a", title='Line Chart with Plotly Express')
st.plotly_chart(fig2)

st.markdown("----")
st.header("Session State Examples")

# Example 1: Counter
st.subheader("Counter Example")
if 'count' not in st.session_state:
    st.session_state.count = 0

st.write('Current count:', st.session_state.count)

if st.button('Increment Count'):
    st.session_state.count += 1
    st.experimental_rerun()

# Example 2: List of items
st.subheader("List of Items Example")
if 'items' not in st.session_state:
    st.session_state.items = []

new_item = st.text_input('Enter a new item:')
if st.button('Add Item') and new_item:
    st.session_state.items.append(new_item)
    st.session_state.new_item_key = "" # Clear the text input after adding
    st.experimental_rerun()

st.write('Items in the list:', st.session_state.items)
```
