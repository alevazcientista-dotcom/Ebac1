````python
# Juntando todos os DataFrames em um só
df_vestibular = pd.concat([
    df_matematica, df_portugues, df_geografia,
    df_ingles, df_historia, df_fisica, df_quimica
], axis=1)

df_vestibular.head()
````

**3 - Percentual Geral de cada aluno**

````python
#Percentual geral de cada aluno
total_questoes = sum(numero_questoes.values())

df_vestibular['Percentual_geral'] = (df_vestibular['Total_de_acertos'] / total_questoes) * 100
df_vestibular.head()
````

**4 - Quantos alunos tiveram nota maior que 45**

````python
#Quantos alunos tiveram nota maior que 45
alunos_aprovados = df_vestibular[df_vestibular['Total_de_acertos'] > 45]
quantidade_aprovados = len(alunos_aprovados)

print(f"Número de alunos com nota maior que 45: {quantidade_aprovados}")
````

**Passo 2: Criar uma lista das colunas desse DataFrame e escolher as que são referentes ao estabelecimento.**

````python
# 2. Criar uma lista das colunas desse DataFrame e escolher as que são referentes ao estabelecimento.
colunas_df_menores_18 = df_menores_18.columns
colunas_estabelecimento = [col for col in colunas_df_menores_18 if 'estabelecimento' in col.lower() or 'estalecimento' in col.lower()]

print("Colunas referentes ao estabelecimento:", colunas_estabelecimento)
````

**Passo 3: Fazer uma contagem do campo `vacina_categoria_nome` (exploração adicional).**

**Passo 4: Selecionar somente as variáveis referentes ao estabelecimento usando o método `.loc` e aplicar `.drop_duplicates`.**

````python
# 4. Usando o método .loc, selecione somente essas variáveis e aplique o método .drop_duplicates.
df_estabelecimentos_unicos = df_menores_18.loc[:, colunas_estabelecimento].drop_duplicates().reset_index(drop=True)
display(df_estabelecimentos_unicos.head())
````

**Passo 5: Contar a quantidade de vacinas aplicadas a menores de 18 anos por estabelecimento.**

````python
# Contar a quantidade de vacinas aplicadas a menores de 18 anos por estabelecimento.
contagem_por_estabelecimento = df_menores_18['estabelecimento_valor'].value_counts().reset_index()
contagem_por_estabelecimento.columns = ['estabelecimento_valor', 'quantidade_vacinas_menores_18']
display(contagem_por_estabelecimento.head())
````

````python
# Combinar as informações dos estabelecimentos únicos com a contagem de vacinas.
df_estabelecimentos_com_contagem = pd.merge(
    df_estabelecimentos_unicos,
    contagem_por_estabelecimento,
    on='estabelecimento_valor',
    how='left'
)
display(df_estabelecimentos_com_contagem.head())
````
