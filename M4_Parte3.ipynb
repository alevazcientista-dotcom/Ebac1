**1 - Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), e que devolve os dados da consulta no formato DataFrame.**

**2 - Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?**

````python
import pandas as pd
import requests
from typing import Optional

def consultar_pvl(uf: str, tipo_interessado: str) -> Optional[pd.DataFrame]:
    """
    Consulta a API do Tesouro Nacional (PVL) para um determinado estado (UF)
    e tipo de interessado ('Estado' ou 'Município').

    Args:
        uf: A sigla da Unidade Federativa (ex: 'RJ', 'SP').
        tipo_interessado: O tipo de interessado ('Estado' ou 'Município').

    Returns:
        Um DataFrame do pandas contendo os dados filtrados, ou None em caso de erro na requisição.
    """
    endereco_url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
    try:
        resposta = requests.get(endereco_url)
        resposta.raise_for_status() # Lança um erro para status codes ruins (4xx ou 5xx)
        dados = resposta.json()
        df_dados = pd.DataFrame(dados['items'])
        filtro_dados = df_dados[(df_dados['uf'] == uf) & (df_dados['tipo_interessado'] == tipo_interessado)].copy()
        return filtro_dados
    except requests.exceptions.RequestException as erro_requisicao:
        print(f"Erro ao fazer a requisição para a API: {erro_requisicao}")
        return None
    except KeyError:
        print("Erro ao processar os dados da API. A chave 'items' não foi encontrada.")
        return None
    except Exception as erro_inesperado:
        print(f"Ocorreu um erro inesperado: {erro_inesperado}")
        return None
````

````python
# 2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?
df_mg_estado = consultar_pvl(uf='MG', tipo_interessado='Estado')

if df_mg_estado is not None:
    solicitacoes_arquivadas_mg = df_mg_estado[df_mg_estado['status'] == 'Arquivado por decurso de prazo']
    numero_solicitacoes = len(solicitacoes_arquivadas_mg)
    print(f"Número de solicitações para o Estado de Minas Gerais com status 'Arquivado por decurso de prazo': {numero_solicitacoes}")
else:
    print("Não foi possível obter os dados para Minas Gerais (Estado).")
````

````python
# 3. Qual é o município da Bahia com mais solicitações deferidas?
df_ba_municipio = consultar_pvl(uf='BA', tipo_interessado='Município')

if df_ba_municipio is not None:
    solicitacoes_deferidas_ba = df_ba_municipio[df_ba_municipio['status'] == 'Deferido']
    municipio_mais_deferidas = solicitacoes_deferidas_ba['interessado'].value_counts().head(1)
    if not municipio_mais_deferidas.empty:
        municipio = municipio_mais_deferidas.index[0]
        quantidade = municipio_mais_deferidas.values[0]
        print(f"O município da Bahia com mais solicitações deferidas é {municipio} com {quantidade} solicitações.")
    else:
        print("Não há solicitações deferidas para municípios na Bahia.")
else:
    print("Não foi possível obter os dados para a Bahia (Município).")
````

````python
# 4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
df_ba_estado = consultar_pvl(uf='BA', tipo_interessado='Estado')

if df_ba_estado is not None:
    nome_arquivo = 'solicitacoes_bahia_estado.csv'
    df_ba_estado.to_csv(nome_arquivo, index=False)
    print(f"Dados de solicitações da Bahia (Estado) salvos em '{nome_arquivo}'")
else:
    print("Não foi possível obter os dados para a Bahia (Estado) para salvar o arquivo.")
````
